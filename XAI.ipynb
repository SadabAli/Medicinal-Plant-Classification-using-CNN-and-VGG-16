```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from lime import lime_image
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Load the Keras model
model = tf.keras.models.load_model(r'D:\MedicalPlant\VGG16_version_2.keras')

# Define class labels
class_labels = [
    'Arive-Dantu', 'Basale', 'Betel', 'Crape_Jasmine', 'Curry', 'Drumstick', 
    'Fenugreek', 'Guava', 'Hibiscus', 'Indian_Beech', 'Indian_Mustard', 
    'Jackfruit', 'Jamaica_Cherry-Gasagase', 'Jamun', 'Jasmine', 'Karanda', 
    'Lemon', 'Mango', 'Mexican_Mint', 'Mint', 'Neem', 'Oleander', 'Parijata', 
    'Peepal', 'Pomegranate', 'Rasna', 'Rose_apple', 'Roxburgh_fig', 
    'Sandalwood', 'Tulsi'
]

# Preprocess the image (same as Flask)
def preprocess_image(image_path):
    # Load and resize the image
    image = load_img(image_path, target_size=(224, 224))
    # Convert to array and normalize
    image = img_to_array(image) / 255.0
    # Add batch dimension
    image = np.expand_dims(image, axis=0)
    return image

# LIME explanation
def explain_image(image_path):
    # Preprocess the image
    image = preprocess_image(image_path)

    # Perform prediction
    predictions = model.predict(image)
    predicted_index = np.argmax(predictions)
    predicted_label = class_labels[predicted_index]
    predicted_confidence = predictions[0][predicted_index] * 100

    print(f"Predicted Class: {predicted_label}")
    print(f"Confidence: {predicted_confidence:.2f}%")

    # Initialize LIME explainer
    explainer = lime_image.LimeImageExplainer()

    # Define prediction function for LIME
    def predict_fn(images):
        return model.predict(images)

    # Generate explanation
    explanation = explainer.explain_instance(
        image=image[0],  # Pass the image without batch dimension
        classifier_fn=predict_fn,  # Pass the prediction function
        top_labels=5,  # Number of top classes to explain
        num_samples=2000  # Number of perturbed samples
    )

    # Get visualization for the predicted label
    temp, mask = explanation.get_image_and_mask(
        label=predicted_index,
        positive_only=True,  # Highlight regions contributing positively
        hide_rest=False,  # Do not hide other regions
        num_features=10,  # Number of superpixels to display
        min_weight=0.01  # Minimum importance weight
    )

    # Visualize the explanation
    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.imshow(image[0])  # Original image
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(mark_boundaries(temp, mask))  # Explanation with superpixels
    plt.title(f"Explanation for '{predicted_label}'")
    plt.axis('off')

    plt.show()

# Main function
if __name__ == "__main__":
    test_image_path = r'D:\MedicalPlant\new_train\Jamun\__3_5719190.png'
    
    # Explain the image
    explain_image(test_image_path)
```
